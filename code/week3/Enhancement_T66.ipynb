```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## Custom define histogram function

```python
def histogram(img):
    h = np.zeros([256,1], dtype = np.uint16)
    r,c = img.shape
    for i in range(r):
        for j in range(c):
            h[img[i,j]] = h[img[i,j]] + 1
    return h
```

```python
img = cv2.imread('./images/cameraman.tif', 0)
```

## matplotlib show image

```python
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.title("grayscale image")
  
```

## cv2 show image

```python
cv2.imshow('original image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
h = histogram(img)
```

## Single frame plot

```python
plt.plot(h)
plt.title('Original histogram') # original histogram
```

## Multiple plots

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.title("grayscale image")

fig.add_subplot(1, 2, 2)
plt.plot(h, color='b')
plt.title('Histogram for gray scale image')
plt.show()
```

## Create histogram from cv2

```python
h = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(h)
plt.title('Original histogram') # original histogram
```

## Create histogram from matplotlib

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

## Histogram of Color image

```python
img = cv2.imread('./images/mandril_color.tif')
```

```python
hb = histogram(img[:,:,0])
hg = histogram(img[:,:,1])
hr = histogram(img[:,:,2])
```

```python
fig = plt.figure()
fig.add_subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("color image")

fig.add_subplot(2, 2, 2)
plt.plot(hb, color='b')
plt.title('Histogram for blue image')
plt.show()

fig.add_subplot(2, 2, 3)
plt.plot(hg, color='g')
plt.title('Histogram for green image')
plt.show()

fig.add_subplot(2, 2, 4)
plt.plot(hr, color='r')
plt.title('Histogram for red image')
plt.show()
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("color image")

r, c, n = img.shape 
color_fig = ['b', 'g', 'r']
fig.add_subplot(1, 2, 2)
for i in range(n):
    hist= cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color= color_fig[i])
plt.title('Histogram for gray scale image')
plt.show()
```
