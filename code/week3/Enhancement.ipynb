```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def gamma_correction(img, gamma):
    img = img.astype(np.float64)/np.amax(img)
    gamma_img = img**gamma
    return np.array(255*gamma_img, dtype = 'uint8')
```

```python
#fp = "./images/A5.tif"
fp = "./images/dark2.jpg"
img = cv2.imread(fp, 0)
res = gamma_correction(img, 0.2)
mul_img = cv2.hconcat([img, res])
# cv2.imshow("color image",mul_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(mul_img, cmap='gray')
plt.axis('off')

#cv2.imwrite('./outs/A5_enhance.png',res.astype('uint8'))
#cv2.imwrite('./outs/dark_enhance.jpg',res.astype('uint8'))
```
