```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ShowImage(img, gray):
    plt.axis("off")
    if gray:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    plt.show()
```

```python
def bgr2rgb(im):
    b = im[:,:,0]
    g = im[:,:,1]
    r = im[:,:,2]
    return cv2.merge((r, g, b))
```

```python
bgr = cv2.imread('./images/redgreen.png')
```

```python
rgb = bgr2rgb(bgr)
```

```python
#rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
```

```python
r,g,b = cv2.split(rgb)
```

```python
ShowImage(r, True)
```
