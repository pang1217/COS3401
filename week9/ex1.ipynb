```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
def plt_show(im):
    plt.axis("off")
    plt.imshow(im, cmap="gray")
    plt.show()
```

```python
def flip(img):
    r, c = img.shape
    out = np.zeros_like(img, dtype='uint8')
    for i in range(r):
        for j in range(c):
            # out[r-i-1, c-j-1] = img[i,j]
            # out[r-i-1, j] = img[i,j]
            out[i, c-j-1] = img[i,j]
    return out.astype(np.uint8)
```
