```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_sub_image(img):
    h, w = img.shape
    w_cutoff = w // 2
    img_left = img[:,:w_cutoff]
    img_right = img[:,w_cutoff:]
    return img_left, img_right 
```

```python
def plt_show(im):
    plt.axis("off")
    plt.imshow(im, cmap="gray")
    plt.show()
```

## Main Processing, First is to get image array. 

```python
img = cv2.imread('./images/gray3.png', 0)
plt_show(img)
```

```python
sub1, sub2 = split_sub_image(img)
```

```python
plt_show(sub1)
```

```python
plt_show(sub2)
```

```python
sub11, sub12 = split_sub_image(sub1.transpose())
sub21, sub22 = split_sub_image(sub2.transpose())
```

```python
sub11, sub12 = sub11.transpose(), sub12.transpose()
sub21, sub22 = sub21.transpose(), sub22.transpose()
```

```python
plt_show(sub11)
```

```python
plt_show(sub22)
```
