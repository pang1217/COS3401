```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
def cv_show(im,msg):
    cv2.imshow(msg,im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()	
```

```python
def plt_show(im):
    plt.axis("off")
    plt.imshow(im, cmap="gray")
    plt.show()
    
```

```python
def hist_img_v0(f):
    m, n = f.shape
    h = [0.0] * 256
    for i in range(m):
        for j in range(n):
            h[f[i, j]]+=1
    return np.array(h, dtype='uint16')
```

```python
img = cv2.imread("./images/test.png", 0)
hist = hist_img_v0(img)
plt.plot(hist)
plt.show()

```

```python
out = img.copy()
thresh = 100
out[out<thresh] = 0
out[out>thresh] = 255
```
