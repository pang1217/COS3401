```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ShowImage(img):
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
def flip_image(img):
    r,c = img.shape
    out = np.zeros_like(img, dtype='u1') 
    for i in range(r):
        for j in range(c):
            out[r-i-1,c-j-1] = img[i,j]
            #out[r-i-1,j] = img[i,j]
            #out[i,c-j-1] = img[i,j]
    return out.astype(np.uint8)
```

```python
img = cv2.imread("./images/cameraman.png", 0)
```

```python
flip_img = flip_image(img)
```

```python
mul_img = cv2.hconcat([img, flip_img])
ShowImage(mul_img)
```
