```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ShowImage(img):
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
def scale_image(img, S):
    r,c = img.shape
    out = np.zeros_like(img, dtype='u1') 
    Sx = S[0]
    Sy = S[1]
    S = np.array([[Sx,0, 0],[0, Sy, 0]])
    for i in range(r):
        for j in range(c):
            x = j # column
            y = i # row
            xy = np.array([x, y, 1])
            new_xy = np.dot(S, xy)  
            xn = int(new_xy[0])
            yn = int(new_xy[1])
            if 0 <= xn < c and 0 <= yn < r:
                out[yn,xn] = img[y,x]
    return out.astype(np.uint8)
```

```python
img = cv2.imread("./images/cameraman.png", 0)
```

```python
scales = (0.6, 0.6)      # (sx,sy)
scale_img = scale_image(img, scales)
```

```python
mul_img = cv2.hconcat([img, scale_img])
ShowImage(mul_img)
```
