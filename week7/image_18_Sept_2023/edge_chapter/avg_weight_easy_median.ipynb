```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("./images/lena_noise_512.png", 0)
```

```python
plt.axis("off")
plt.imshow(img, cmap='gray')
plt.show()
```

```python
sz = 11
bd = sz//2
print(bd)
```

```python
out = np.zeros_like(img, dtype = 'uint8')
```

```python
(m,n) = img.shape
num_ele = sz*sz
inx = num_ele//2
```

```python

for i in range(bd,m-bd):
    for j in range(bd,n-bd):
        a = img[i - bd : i + bd + 1, j - bd : j + bd + 1]
        b = np.ravel(a)
        c = np.sort(b)
        out[i,j] = c[inx] 
```

```python
imgs = cv2.hconcat([img, out])
```

```python
plt.axis("off")
plt.imshow(imgs, cmap='gray')
#plt.show()
plt.savefig('./outs/filtered.png')
```
