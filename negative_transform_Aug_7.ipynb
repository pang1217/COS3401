```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Input GrayScale method 1

```python
img1 = cv2.imread('./images/lena_color_256.png', 0)
```

```python
img = cv2.imread('./images/lena_color_256.png')
```

## Input GrayScale method 2

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

## Input GrayScale method 3

```python
img = img.astype(np.float16)
b = 0.1140*img[:,:,0]
g = 0.5870*img[:,:,1]
r = 0.2989*img[:,:,2]
img3 = b + g + r
img3 = img3.astype(np.uint8)
```

```python
fig = plt.figure()
fig.add_subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 3, 3)
plt.imshow(img3, cmap='gray')
plt.axis('off')

plt.savefig('./outs/fig.png')
```

```python
row, col, spaces = img.shape
print(row, col, spaces)
```

```python
img4 = np.zeros([row, col], dtype='float16')
```

```python
for i in range(row):
    for j in range(col):
        img4[i,j] = 0.1140*img[i,j,0] + 0.5870*img[i,j,1] + 0.2989*img[i,j,2]
```

```python
img4 = img4.astype(np.uint8)
```

```python
fig = plt.figure()
fig.add_subplot(2, 2, 1)
plt.imshow(img1, cmap='gray')
plt.axis('off')

fig.add_subplot(2, 2, 2)
plt.imshow(img2, cmap='gray')
plt.axis('off')

fig.add_subplot(2, 2, 3)
plt.imshow(img3, cmap='gray')
plt.axis('off')

fig.add_subplot(2, 2, 4)
plt.imshow(img4, cmap='gray')
plt.axis('off')

plt.savefig('./outs/fig_imgs4.png')
```

# Negative Transform

```python
neg1 = 255 - img1
```

```python
neg2 = np.zeros_like(img1, dtype='uint8')
```

```python
for i in range(row):
    for j in range(col):
        neg2[i,j] = 255 - img1[i,j]
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(neg1, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 2, 2)
plt.imshow(neg2, cmap='gray')
plt.axis('off')

plt.savefig('./outs/fig_negs.png')
```

```python
cv2.imshow('Negative Image',neg2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(neg1.shape, neg2.shape)
```

```python
neg3 = cv2.hconcat([neg1, neg2])
print(neg3.shape)
```

```python
cv2.imshow('Negative Image',neg3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
neg4 = cv2.vconcat([neg1, neg2])
print(neg4.shape)
```

```python
cv2.imshow('Negative Image',neg4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
