```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def histogram (img):
    h = np.zeros([256, 1], dtype = np.uint16)
    r, c = img.shape
    for i in range(r):
        for j in range(c):
            h[img[i, j]] = h[img[i, j]] + 1
            
    return h
```

```python
img = cv2.imread('./images/cameraman2.jpg', 0) # option 0 is greyscale
```

```python
plt.imshow(img, cmap = 'gray') #support colors
plt.axis('off')
plt.title('grayscale image')
```

```python
cv2.imshow('greyscale', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
h = histogram(img)
fig = plt.figure()
plt.plot(h)
plt.title('Original Histogram')
```

## create histogram via cv2

```python
h = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(h)
plt.title('Original Histogram')
```

```python
#Show histogram of gray level image
#  2nd method
plt.hist(img.ravel(), 256, [0,256])
plt.title('Histogram for gray scale image')
plt.show()
```

```python
img = cv2.imread('./images/mandril_color.png', 1)
h_blue = histogram(img[:, :, 0])
h_green = histogram(img[:, :, 1])
h_red = histogram(img[:, :, 2])
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('color image')

fig.add_subplot(1, 2, 2)
plt.imshow(h_blue, color='b')
plt.show()

fig.add_subplot(2, 2, 2)
plt.imshow(h_green, color='g')
plt.show()

fig.add_subplot(2, 2, 2)
plt.imshow(h_red, color='r')
plt.show()
```
