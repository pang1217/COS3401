```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
def imshow(img):
    plt.axis('off')
    plt.imshow(img, cmap='gray')
```

```python
def cv2_show(caption, im):
    cv2.imshow(caption,im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def gamma_correction(img, gamma):
    img = img.astype(np.float64)/np.amax(img)
    gamma_img = img**gamma
    return np.array(255*gamma_img, dtype = 'uint8')
```

```python
img = cv2.imread('./images/ivus.png', 0)
```

```python
imshow(img)
```

```python
im = gamma_correction(img, 2.8)
imshow(im)
```

```python
result = cv2.hconcat([img, im])
imshow(result)
plt.savefig('./outs/hw3.png')
```
