```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def log_function(img) :
    im = img.astype(np.float16)
    c = 255/np.log(1 + np.max(im))
    log_img = c * np.log(1 + img)
    return log_img.astype(np.uint8)
```

```python
def cv2_show(caption, img):
    cv2.imshow(caption, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def plt_show (im):
    fig = plt.figure()
    plt.imshow(im, cmap='gray')
    plt.axis('off')
    plt.savefig('./outs/enchace.png')
```

```python
def log_manual(img) :
    row, col = img.shape
    im = img.astype(np.float16)
    log_img = np.zeros_like(img, dtype='float16')
    c = 255/np.log(1 + np.max(im))
    for i in range (row) :
        for j in range (col) :
            log_img[i][j] = c * np.log(1 + img[i, j])
    return log_img.astype(np.uint8)
```

```python
img = cv2.imread('./images/dark2.jpg', 0)
```

```python
im = log_function(img)
mul_im = cv2.hconcat([img, im])
```

```python
cv2_show('enchance img', mul_im)
```

```python
plt_show(mul_im)
```

```python
im = log_manual(img)
mul_im = cv2.hconcat([img, im])
```

```python
cv2_show('enchance img', mul_im)
```
