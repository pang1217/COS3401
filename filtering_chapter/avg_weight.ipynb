```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Read Image

```python
f = cv2.imread("./images/lena_noise_512.png", 0)
```

## Show Image

```python
print(type(f), f.shape)
```

```python
plt.axis("off")
plt.imshow(f, cmap='gray')
plt.show()
```

## Get Image Information 

```python
r, c = f.shape
print(r,c)
```

```python
fs = np.zeros_like(f, dtype='float32')
f = f.astype(np.float32)
```

```python
print(type(fs), fs.shape)
```

```python
for i in range (1, r-1):
    for j in range(1, c-1):
        temp = f[i-1, j-1] + f[i-1, j] + f[i-1, j + 1] + f[i, j-1] + f[i, j] + f[i, j + 1] + f[i + 1, j-1] + f[i + 1, j] + f[i + 1, j + 1] 
        fs[i,j] = temp/9.0

fs = fs.astype(np.uint8)
f = f.astype(np.uint8)
```

```python
img = cv2.hconcat([f, fs])
plt.axis("off")
plt.imshow(img, cmap='gray')
plt.show()
```
